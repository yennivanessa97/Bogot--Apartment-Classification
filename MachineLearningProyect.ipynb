```python
import re

def is_legible(text):
    if not text.strip():  # Verifica si el texto está vacío o contiene solo espacios
        return False
    if len(re.findall(r'[a-zA-Z0-9]', text)) < 10 or len(text) < 20:
        return False
    if re.search(r'[^\x00-\x7F]+', text):
        return False
    return True

# Aplicar la función para verificar la legibilidad
df_copia['legible'] = df_copia['descripcion'].apply(lambda x: is_legible(str(x)) if pd.notnull(x) else False)

# Contar y mostrar documentos ilegibles
documentos_ilegibles = df_copia[~df_copia['legible']].shape[0]
print(f"Número de documentos ilegibles: {documentos_ilegibles}")

# Mostrar ejemplos de documentos ilegibles
ejemplos_ilegibles = df_copia[~df_copia['legible']]['descripcion']
print(ejemplos_ilegibles)
```

**LIMPIEZA CON REGEX**

```python
import re
```

```python
# Permitir letras minúsculas, dígitos y espacios

pat = re.compile(r"[^a-z0-9 ]")
```

```python
clean_text = re.sub(pat, "", lower_text)
spaces = re.compile(r"\s{2,}")
spaces_text = re.sub(spaces, " ", clean_text)
print(spaces_text)
```

```python
#función con todo el proceso

pat = re.compile(r"[^a-z0-9 ]")
spaces = re.compile(r"\s{2,}")

def preprocess(text, min_len=1, max_len=23):

  #crear un documento de spacy a partir del texto
    doc = nlp(text)

  #eliminar stopwords
    filtered_tokens = filter(
        lambda token: not token.is_stop,
        doc
        )
  #filtrar palabras por longitud
    filtered_tokens2 = filter(
        lambda token: len(token) >= min_len and len(token) <= max_len,
            filtered_tokens
        )
  #Obtener los lemmas de cada token
    lemmas = map(
        lambda token: token.lemma_,
        filtered_tokens2
        )
  #unir los lemmas en un solo texto
    lemma_text = " ".join(lemmas)

  #normalizar el texto para eliminar caracteres especiales
    norm_text = unidecode(lemma_text)

  #convertir el texto a minusculas
    lower_text = norm_text.lower()

  #Eliminar caracteres especiales que no sean letras,numeros o textos
    clean_text = re.sub(pat, "", lower_text)

  #Eliminar espacios duplicados
    spaces_text = re.sub(spaces, " ", clean_text)

  #Retornar el texto limpio y sin espacios innecesarios
    return spaces_text.strip()


```

```python
#ejemplo en un texto del dataframe
text = df_copia['descripcion'].iloc[0]
processed_text = preprocess(text)
print(processed_text)
```

```python
#Uso del método apply para aplicar el preprocesamiento sobre todo el corpus
df_copia['descripcion'] = df_copia['descripcion'].fillna('').astype(str)
pre_corpus = df_copia['descripcion'].apply(preprocess).tolist()
print(pre_corpus[:10])
```

En el preprocesamiento del texto,primero se utilizó Spacy para poder hacer la divición de los tokens,eliminando stopwords y filtrando por longitud de palabras, luego se generaron los lemmas de cada token para obtener una versión más simplicada de las palabras.Además, se normalizó el texto eliminando tildes y otros caracteres especiales con la librería unidecode y se convirtió el texto a minúsculas. Finalmente, se aplicaron expresiones regulares para eliminar cualquier carácter que no fuera letra minúscula, número o espacio, y se eliminaron los espacios duplicados.
